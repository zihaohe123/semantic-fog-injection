## Convert the AdvBench dataset from huggingface into Llama-factory-compatible format.

```python
from datasets import load_dataset
```

```python
dataset = load_dataset("walledai/AdvBench")
```

```python
dataset
```

```python
dataset['train']['prompt']
```

```python
dataset['train']['target']
```

```python
import json
```

```python
# Convert to the required JSON format
data = []
for prompt, output in zip(dataset['train']['prompt'], dataset['train']['target']):
    data.append({
        "instruction": prompt,
        "input": "",
        "output": output
    })

# Save to a JSON file
with open("adv_bench.json", "w") as f:
    json.dump(data, f, indent=4)

print("JSON file created successfully!")
```
